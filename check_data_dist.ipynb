```python
import os
import torch
import numpy as np
from data import Task1Data, Task2Data, Task3Data
import os
import statsmodels.api as sm
import matplotlib.pyplot as plt

norm_dataset = Task1Data()
```

```python
roi = 0
check_data = norm_dataset.ct_feas[...,0][...,roi]
```

```python
check_data.shape
```

```python
demos = dict()
for sex in range(2):
    for age in range(100):
        use_data = check_data[(norm_dataset.sexs == sex) & (norm_dataset.ages == age)]
        mean_ = np.mean(use_data)
        var_ = np.std(use_data)
        count_ = len(use_data)

        demos[sex,age] = (mean_, var_, count_)
```

```python
to_plot_mean = [demos[(0,i)][0]-0.4 for i in range(100)]
to_plot_var = [demos[(0,i)][1] for i in range(100)]
to_plot_count = [demos[(0,i)][2]/1000 for i in range(100)]

plt.plot(range(100), to_plot_mean,label='mean')
plt.bar(range(100), to_plot_var,label = 'var',alpha=0.4)
plt.bar(range(100), to_plot_count,alpha=0.4)

plt.legend()
```
